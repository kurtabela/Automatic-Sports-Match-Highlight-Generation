### Helper functions

```python
def get_sec(time_str):
    """Get Seconds from time."""
    if len(time_str.split(':')) == 2:
        m, s = time_str.split(':')
        h = 0
    else:
        h, m, s = time_str.split(':')
    return int(h) * 3600 + int(m) * 60 + int(s)
```

### Adjust timestamps on the transcripts
This is done to reflect the change in the length of the file since the video is split

```python
import copy
def adjustFirstAndMiddleBitFromAnnotations(speechToTextFile, beginningTimeToAdjust, endTimeToAdjust, adjustBy):
  toWrite = []
  with open(speechToTextFile, encoding='utf-8', errors='ignore') as json_file: 
    transcript = json.load(json_file)
    toWrite = copy.deepcopy(transcript)
    for i, sentence in enumerate(transcript):
      print(sentence)
      if sentence['start'] >= beginningTimeToAdjust and sentence['start'] <= endTimeToAdjust:
        print("adjusting start")
        sentence['start'] -= adjustBy 
        toWrite[i] = sentence
  writeToFile(speechToTextFile, toWrite)



def removeFirstAndMiddleBitFromAnnotations(speechToTextFile, fromSeconds, endSeconds):
 
  with open(speechToTextFile, encoding='utf-8', errors='ignore') as json_file: 
    transcript = json.load(json_file)
    toWrite = copy.deepcopy(transcript)
    for sentence in transcript:
      print(sentence)
      if sentence['start'] >= fromSeconds and sentence['start'] <= endSeconds:
        toWrite.remove(sentence)
  writeToFile(speechToTextFile, toWrite)


def writeToFile(speechToTextFile, toWrite):
  with open(speechToTextFile, 'w') as open_file:
    json.dump(toWrite, open_file)
```

```python
# Copy the transcript into first half and second half
from shutil import copyfile
pathToTranscripts = "/content/data/MyDrive/Thesis/Datasets/Transcripts/"
firstHalfTranscript = pathToTranscripts + fileName.partition(".mp4")[0] + "_FirstHalf_TEST.json"
secondHalfTranscript = pathToTranscripts + fileName.partition(".mp4")[0] + "_SecondHalf_TEST.json"
copyfile(rawTranscripts, firstHalfTranscript)
copyfile(rawTranscripts, secondHalfTranscript)


# Read in the file
with open(firstHalfTranscript, 'r', encoding='utf-8', errors='ignore') as file :
  filedata = file.read()

# Replace the target string
filedata = filedata.replace('"', "'")
filedata = filedata.replace("'text': '", '"text": "')
filedata = filedata.replace("', 'start':", '", "start":')
filedata = filedata.replace(", 'duration':", ', "duration":')


# Write the file out again
with open(firstHalfTranscript, 'w', encoding='utf-8', errors='ignore') as file:
  file.write(filedata)


# Read in the file
with open(secondHalfTranscript, 'r', encoding='utf-8', errors='ignore') as file :
  filedata = file.read()

# Replace the target string
filedata = filedata.replace('"', "'")
filedata = filedata.replace("'text': '", '"text": "')
filedata = filedata.replace("', 'start':", '", "start":')
filedata = filedata.replace(", 'duration':", ', "duration":')


# Write the file out again
with open(secondHalfTranscript, 'w', encoding='utf-8', errors='ignore') as file:
  file.write(filedata)
```

```python
# For first half

# Remove beginning
removeFirstAndMiddleBitFromAnnotations(firstHalfTranscript, 0, toRemoveInTheBeginning)

# Remove half time till the end
removeFirstAndMiddleBitFromAnnotations(firstHalfTranscript, beginningOfHalfTime , 999999)

# Adjust first half
adjustFirstAndMiddleBitFromAnnotations(firstHalfTranscript, 0, beginningOfHalfTime, toRemoveInTheBeginning)


# For second half

# Remove half time and earlier
removeFirstAndMiddleBitFromAnnotations(secondHalfTranscript, 0, endOfHalfTime)

# Remove ending
removeFirstAndMiddleBitFromAnnotations(secondHalfTranscript, endOfGame, 99999)

# Adjust second half
adjustFirstAndMiddleBitFromAnnotations(secondHalfTranscript, endOfHalfTime-1, endOfGame, endOfHalfTime)
```
